```python
# to itirate the same formula to this data sets column "Date/Time"
data['Date/Time'] = data['Date/Time'].map(pandas.to_datetime)
```

```python
data.head()
```

```python
# start analysis with the defining the function

# this will get us the day of the week
def get_dom(dt):
    return dt.day

data['dom']=data['Date/Time'].map(get_dom)
```

```python
data.tail()
```

```python
data.head()
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday']=data['Date/Time'].map(get_weekday)
```

```python
data.head()
```

```python
def get_hour(dt):
    return dt.hour
data['hour'] = data['Date/Time'].map(get_hour)
```

```python
data.head()
```

```python
data.tail()
```

# now start data analysis

```python
# just checking
hist(data.dom)
```

```python
hist(data.dom, bins =30, rwidth = .8, range = (0.5,30.5))
xlabel('days of the month')
ylabel('frequency')
title('frequency vs days')
```

```python
def count_rows(rows):
    return len(rows)

data01 = data.groupby('dom').apply(count_rows)
```

```python
data01
```

```python
# plot the above data01
plot(data01)
```

```python
data01_sort_values= data01.sort_values()
```

```python
data01_sort_values
```

```python
bar(range(1,31), data01_sort_values)
xticks(range(1,31), data01_sort_values.index)
xlabel('days of the month')
ylabel('frequency')
title('frequency vs days')
```
